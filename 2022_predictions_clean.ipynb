```python
import pandas as pd
```

```python
df1 = pd.read_csv("data/candidate_summary_2022.csv")
df_2022 = df1.copy()
df_2022
```

```python
cols = df_2022.columns
```

```python
df_2022.info()
```

```python
df_2022 ['raised_from_pacs'] = df_2022['Other_Committee_Contribution'] + df_2022['Party_Committee_Contribution'] + df_2022['Transfer_From_Other_Auth_Committee'] 
```

```python
df_2022 = df_2022.rename(columns={"Cand_Name": "first_last_party", "Cand_Party_Affiliation": "party", "Cand_Incumbent_Challenger_Open_Seat": "crpico", "Individual_Contribution": "raised_from_individuals", "Total_Contribution": "raised_total", "Individual_Unitemized_Contribution": "raised_unitemized"})
```

```python
df_2022.columns
```

```python
cols_keep = ['first_last_party', 'Cand_Office_St', 'Cand_Office', 'party', 'crpico', 'raised_from_pacs', 'raised_from_individuals', 'raised_total', 'raised_unitemized']
df_2022 = df_2022[cols_keep]
df_2022.sample(5)
```

```python
df_2022['Cand_Office'] = df_2022['Cand_Office'].replace({'H': 'House','S': 'Senate'})
```

```python
rep = {
    'REP': 'R',
    'DEM': 'D',
    'LIB': 'L',
    'IND': 'I',
    'N': 'U',
    'CHALLENGER': 'C',
    'OPEN': 'O',
    'INCUMBENT': 'I',    
}
df_2022 = df_2022.replace(rep)
df_2022.sample(5)
```

```python
for col in df_2022.columns:
    print(df_2022[col].value_counts())
```

```python
# list of values to replace if not in
X = ['R','D','L','I','U']
# replace all values in row not in X with '3'
df_2022.loc[~df_2022.party.isin(X), 'party'] = '3'
```

```python
party = df_2022['party'].value_counts()
party
```

```python
(df_2022['raised_total'] == 0).sum()
```

```python
df_2022.fillna('', inplace=True)
```

```python
df_2022.to_csv('data/cand_summary_2022_clean.csv', index=False)
```
