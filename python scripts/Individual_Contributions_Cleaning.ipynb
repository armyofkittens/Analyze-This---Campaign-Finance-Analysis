```python
# Load individual_contributions.csv
file_to_load = ("individual_contributions.csv")
ic_df = pd.read_csv(file_to_load, low_memory= False)
ic_df.head()
```

```python
ic_df.count()
```

```python
# Drop columns
ic_clean_df = ic_df.drop(columns=['fec_trans_id', 'ult_org', 'real_code', 'street', 'city', 'recip_code', 'type', 'committee_id', 'other_id', 'old_format_employer_occupation', 'microfilm', 'employer', 'source'])
```

```python
# Check that dropped columns worked
ic_clean_df.columns
```

```python
# Create unique_key column
ic_clean_df['unique_key'] = ic_clean_df['cycle'].map(str) + ic_clean_df['recipient_id'].map(str)
ic_clean_df.head()
```

```python
# Reorder columns
ic_clean_df = ic_clean_df[["id", "cycle", "contributor_id", "contributor_name", "unique_key", "recipient_id", "amount", "date", "org_name", "state", "zip", "gender", "occupation"]]
ic_clean_df.head()
```

```python
# Check date datatype
ic_clean_df.info()
```

```python
# Check for Null Values in recipient_id
ic_clean_df["recipient_id"].isnull().sum()
```

```python
# Drop rows where recipient_id is null
ic_clean_df = ic_clean_df.dropna(subset=['recipient_id'])
```

```python
ic_clean_df.count()
```

```python
# Check for duplicates in id column
duplicates = ic_clean_df.id.duplicated().sum()
print(duplicates)
```

```python
# Convert the date to Y-m-d without time
ic_clean_df['date'] = pd.to_datetime(ic_clean_df['date'], format='%Y-%m-%d')
  
# Filter data for most recent 4 years
twelve_sixteen_df = ic_clean_df.loc[(ic_clean_df['date'] >= '2012-01-01')
                     & (ic_clean_df['date'] < '2016-12-31')]
```

```python
# Check that date was converted
twelve_sixteen_df.info()
```

```python
# Check on column numbers/ null values
twelve_sixteen_df.count()
```

```python
twelve_sixteen_df.head()
```

```python
# Filter data from 2008-2012
eight_twelve_df = ic_clean_df.loc[(ic_clean_df['date'] >= '2008-01-01')
                     & (ic_clean_df['date'] < '2012-12-31')]
```

```python
eight_twelve_df.count()
```

```python
# Filter data from 2004-2008
four_eight_df = ic_clean_df.loc[(ic_clean_df['date'] >= '2004-01-01')
                     & (ic_clean_df['date'] < '2008-12-31')]
```

```python
four_eight_df.count()
```

```python
# Save new file
twelve_sixteen_df.to_csv("ind_cont_12_16.csv", encoding='utf-8', index=False)
```
