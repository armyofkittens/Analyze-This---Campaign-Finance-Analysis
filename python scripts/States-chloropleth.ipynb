```python
import pandas as pd
```

```python
df = pd.read_csv("2022 predictions/prediction_results_adjusted.csv")
df.head()
```

```python
states_raised_totals = df.groupby("Cand_Office_St")['raised_total'].sum()
states_raised_totals = pd.DataFrame(states_raised_totals).reset_index()
states_raised_totals = states_raised_totals.rename(columns={"Cand_Office_St": "State", "raised_total": "Total"})
states_raised_totals
```

```python
states_raised_totals.to_csv("2022 predictions/states_raised_totals.csv")
```
